#***Dividing Dependent And Independent Variables***
Before starting model building, it is important to divide the dataset into two parts: independent variables and dependent variable. The independent variables contain a list of those variables on which the concrete quality is dependent, while the dependent variable is the variable that is dependent on the values of other variables.

```python
# independent variable
x = df.drop(['Concrete Compressive Strength'], axis =1)
# dependent variable
y = df['Concrete Compressive Strength']
```

In this program, the independent variables are stored in the list x, while the dependent variable is stored in y. Specifically, x contains the following independent variables: Cement, Blast Furnace, Fly Ash, Water, Super Plasticizer, Coarse Aggregate, Fine Aggregate, and Age. The dependent variable is Concrete Compressive Strength.

#***Splitting the Data***

To split the training and testing parts, we can use the scikit-learn module train_test_split.

```python
# Splitting the Data
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=42)
```

#***Feature Scaling***
We do the scaling of the data for balancing the data points

```python
# Feature Scaling
stand= StandardScaler()
Fit = stand.fit(xtrain)
xtrain_scl = Fit.transform(xtrain)
xtest_scl = Fit.transform(xtest)
```

#***Applying model***
Using a model in a project means putting the model's insights or predictions into practical use to achieve the project's objectives. This may include making real-time decisions, automating specific processes, improving efficiency, and ultimately delivering value to the organization or end-users. It is crucial to continually assess and update the applied model to ensure its accuracy and relevance in changing circumstances.


#***Linear regression***

```python
# Applying Linear Regression
lr=LinearRegression()
fit=lr.fit(xtrain_scl,ytrain)
print('Score of Ridge Regression is:-',lr.score(xtest_scl,ytest))
print('.......................................................')
print('Mean_Sqrd_Error of Ridge Regression is==',mean_squared_error(ytest,lr.predict(xtest_scl)))
print('root_mean_squared error of Ridge Regression is==',np.sqrt(mean_squared_error(ytest,lr.predict(xtest_scl))))
```

#***Ridge Regression***

```python
# Applying Ridge Regression
rd= Ridge(alpha=0.4)
fit_rd=rd.fit(xtrain_scl,ytrain)
print('Score of Ridge Regression is:-',rd.score(xtest_scl,ytest))
print('.......................................................')
print('Mean_Sqrd_Error of Ridge Regression is==',mean_squared_error(ytest,rd.predict(xtest_scl)))
print('root_mean_squared error of Ridge Regression is==',np.sqrt(mean_squared_error(ytest,rd.predict(xtest_scl))))
```

#***Lasso Regression***

```python
# Applying Lasso Regression
ls= Lasso(alpha=0.4)
fit_ls = ls.fit(xtrain_scl,ytrain)
print('Score of Lasso Regression is:-',ls.score(xtest_scl,ytest))
print('.......................................................')
print('mean_sqrd_error of Lasso Regression is==',mean_squared_error(ytest,ls.predict(xtest_scl)))
print('root_mean_squared error of Lasso Regression is==',np.sqrt(mean_squared_error(ytest,lr.predict(xtest_scl))))
```

#***Random Forest Regressor***

```python
# Applying Random Forest Regressor
rnd= RandomForestRegressor(ccp_alpha=0.4)
fit_rnd= rnd.fit(xtrain_scl,ytrain)
print('Score is:-',rnd.score(xtest_scl,ytest))
print('........................................')
print('Mean_Sqrd_Error is==',mean_squared_error(ytest,rnd.predict(xtest_scl)))
print('Root_Mean_Squared Error of is==',np.sqrt(mean_squared_error(ytest,rnd.predict(xtest_scl))))
```

#***K-Nearest Neighbor (KNN)***

```python
# Applying K - Nearest Neighbors Regressor
knn = KNeighborsRegressor(n_neighbors = 5)
fit_knn = knn.fit(xtrain_scl, ytrain)
print('Score is:-',knn.score(xtest_scl,ytest))
print('........................................')
print('Mean_Sqrd_Error is==',mean_squared_error(ytest,knn.predict(xtest_scl)))
print('Root_Mean_Squared Error of is==',np.sqrt(mean_squared_error(ytest,knn.predict(xtest_scl))))
```

#***Support Vector Machine (SVM)***

```python
# Applying Support Vector Machine Regressor
svr = SVR(kernel = 'linear', C = 1.0)
fit_svr = svr.fit(xtrain_scl, ytrain)
print('Score is:-',svr.score(xtest_scl,ytest))
print('........................................')
print('Mean_Sqrd_Error is==',mean_squared_error(ytest,svr.predict(xtest_scl)))
print('Root_Mean_Squared Error of is==',np.sqrt(mean_squared_error(ytest,svr.predict(xtest_scl))))
```
